```python
from typing import TypedDict, Dict
from langgraph.graph import StateGraph
import math

class AgentState(TypedDict):
     values: list[int]
     name:str
     operation:str
     result: str
```

```python
def perform_operation(state: AgentState):
    """ Perform arithmentic base on the operation value """
    if(state["operation"] == "*"):
       operation_result = math.prod(state["values"])
    elif(state["operation"] == "+"):
        operation_result = sum(state["values"])
    else:
        operation_result = "Unsupported operation"

    state["result"] = f"Hi {state['name']}, Your answer is {operation_result}"
    return state;     
```

```python
graph = StateGraph(AgentState)
graph.add_node("arithmetic_processor", perform_operation)
graph.set_entry_point("arithmetic_processor")
graph.set_finish_point("arithmetic_processor")

app = graph.compile()
```

```python
answer = app.invoke({"values": [1,2,3,4], "name": "Steve", "operation": "*"})
print(answer["result"])
```
