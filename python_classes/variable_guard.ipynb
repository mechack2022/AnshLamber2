class Patient:
    def __init__(self, name):  # Called automatically when creating object
        self.name = name
    
    def __str__(self):         # Called when print() is used
        return f"Patient: {self.name}"

p = Patient("John")
print(p)  # → "Patient: John" (triggers __str__)
